```julia
using JuMP, GLPK

m = Model(GLPK.Optimizer)

@variable(m, x12, Bin)
@variable(m, x13, Bin)
@variable(m, x14, Bin)
@variable(m, x23, Bin)
@variable(m, x25, Bin)
@variable(m, x26, Bin)
@variable(m, x34, Bin)
@variable(m, x35, Bin)
@variable(m, x36, Bin)
@variable(m, x37, Bin)
@variable(m, x46, Bin)
@variable(m, x47, Bin)
@variable(m, x56, Bin)
@variable(m, x58, Bin)
@variable(m, x67, Bin)
@variable(m, x68, Bin)
@variable(m, x78, Bin)



@objective(m, Min, 6x12 + 3x13 + 1x14 + 1x23 + 4x25 + 3x26 + 3x34 + 5x35 + 6x36 + 2x37 + 5x46 + 5x47 + 3x56 + 1x58 
+ 1x67 + 3x68 + 6x78)

@constraint(m, node1source, x12 + x13 + x14 == 1)
@constraint(m, node2, x12 - x23 - x25 - x26 == 0)
@constraint(m, node3, x13 - x34 - x35 - x36 - x37 == 0)
@constraint(m, node4, x14 - x46 - x47 == 0)
@constraint(m, node5, x25 + x35 - x58 == 0)
@constraint(m, node6, x26 + x36 + x46 + x56 - x68 == 0)
@constraint(m, node7, x37 + x47 + x67 - x78 == 0)
@constraint(m, node8sink, x58 + x68 + x78 == 1)

print(m)

JuMP.optimize!(m)

println("Optimal Solution:")
println("x12 = ", JuMP.value(x12))
println("x13 = ", JuMP.value(x13))
println("x14 = ", JuMP.value(x14))
println("x23 = ", JuMP.value(x23))
println("x25 = ", JuMP.value(x25))
println("x26 = ", JuMP.value(x26))
println("x34 = ", JuMP.value(x34))
println("x35 = ", JuMP.value(x35))
println("x36 = ", JuMP.value(x36))
println("x37 = ", JuMP.value(x37))
println("x46 = ", JuMP.value(x46))
println("x47 = ", JuMP.value(x47))
println("x56 = ", JuMP.value(x56))
println("x58 = ", JuMP.value(x58))
println("x67 = ", JuMP.value(x67))
println("x68 = ", JuMP.value(x68))
println("x78 = ", JuMP.value(x78))
println("Objective value: ", JuMP.objective_value(m))
```
